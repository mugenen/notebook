## 動画を埋め込んでみる

### `IPython.display.YouTubeVideo` を使う

```python
import IPython.display
```

```python
IPython.display.YouTubeVideo('me9nj0M46eM')
```

開始時間を指定したりできる。
https://ipython.readthedocs.io/en/stable/api/generated/IPython.display.html#IPython.display.YouTubeVideo

```python
IPython.display.YouTubeVideo('me9nj0M46eM', start=600)
```
